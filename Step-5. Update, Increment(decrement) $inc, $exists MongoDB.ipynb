```python
# Keling hamma dokumentga yosh degan element qo'shib olamiz
for i in range(20):
    current = {"_id":i}
    new_data = {"$set":{"age":i}}
    kolleksiya.update_one(current, new_data)
```

```python
for value in kolleksiya.find():
    print(value)
```

```python
# Endi $inc orqali hamma dokumentdagi yoshlarni 1 yoshga oshiramiz
kolleksiya.update_many({"age":{"$exists":True}}, # $exists bu dokumentda age elementi bor yo'qligini tekshiradi
                       {"$inc":{"age":+1}})
```

```python
for value in kolleksiya.find():
    print(value)
```

## <center> $existis

```python
for value in kolleksiya.find({"age":{"$exists":True}}): # $exists faqat Boolean qiymatni qabul qiladi, ya'ni True yoki False
    print(value)
# Faqat age elementi bor dokumentlarni chiqaradi
```
