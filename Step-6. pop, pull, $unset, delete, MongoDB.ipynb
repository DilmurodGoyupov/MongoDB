## <center> pop - elmentni index bo'yicha o'chiradi

```python
# Arrayning 1-elementini o'chirish. POP
current = {"_id":1}
new_data = {"$pop":{"edu":1}}

kolleksiya.update_one(current, new_data)
```

<center>

# 

## <center> pull - elementni nomi bo'yicha o'chiradi


```python
# Arrayning 0-elementini o'chirish. PULL
current = {"_id":1}
new_data = {"$pull":{"edu":"Tatu"}}

kolleksiya.update_one(current, new_data)
```

## <center> $unset - yangilash, dokumentdan elemntlarni o'chirish uchun ishlatiladi

```python
# Arrayni o'zini ham o'chirb tashlaymiz
kolleksiya.update_one({"edu":[]}, {"$unset":{"edu":""}})
```
