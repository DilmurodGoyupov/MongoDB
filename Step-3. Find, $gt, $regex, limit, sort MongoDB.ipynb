# <center> $gt

```python
query = {"name":{"$gt":"K"}} # bu orqali biz birinchi bo'lib K harfidan boshlanadigan keyin esa alfavit bo'yicha undan keyin keladigan harflar
# bo'yicha qidirishni amalga oshiramiz
# Oddiyroq aytadigan bo'lsak alfavit bo'yicha K harfi va undan keyingi harflardan boshlanadigan name larni topib beradi

for value in kolleksiya.find(query):
    print(value)
# Bizning collectionimizda Elbek, Asal degan name lar bor edi
# pastga e'tbor bersangiz ular chiqmagan, chunki alfavit bo'yicha ular K harfidan oldin keladi, biz esa K harfi va undan keyin turuvchi harflar 
# bo'yicha qidiramiz
```

```python
query = {"name":{"$gt":"Kamron"}}

for value in kolleksiya.find(query):
    print(value)
    
# E'tibor bering hozir Kamron name ga ega bo'lgan dokuent chiqmagan chunki, $gt berilgan simvoldan keyingilarini qdiradi
# "$gt":"Kamron" berdik va u undan keyingilarini qidirishni boshlaydi
```

```python
query = {"age":{"$gt":28}} # bu yerdan biz 28 yoshdan kattalarni qidiramiz

for value in kolleksiya.find(query):
    print(value)
# 28 yoshni o'zi ham chiqmaydi
```

## <center> $gt ni asosan sonlik qiymatlarda ishlatganimiz yaxshi

# <center> $regex

```python
query = {"name":{"$regex":"K"}} # $gt dan farqli olaroq $regex faqt K qatnashgan dokumentlarni topadi. K harfi so'zni o'rtasida kelsa ham topadi

for value in kolleksiya.find(query):
    print(value)
```

```python
query = {"name":{"$regex":"i"}} # $ i harfi qatnashgan hamma dokumentlarni topadi

for value in kolleksiya.find(query):
    print(value)
```

```python
query = {"name":{"$regex":"Alibek"}} # $ i harfi qatnashgan hamma dokumentlarni topadi

for value in kolleksiya.find(query):
    print(value)
```

```python
query = {"age":{"$regex":54}} # $regex parametr sifatida satr qabul qiladi, age bo'lsa bizda int

for value in kolleksiya.find(query):
    print(value)
```

## <center> $regex parametr sifatida satr qabul qiladi, int va boshqa qiymatlarni qabul qilmaydi

## <center> `&gt ni sonli qiymatlarda ishlatsak yaxshiroq bo'ladi(lekin satrli qiymatlarda ham ishlatishimiz mumkin). $regex ni esa faqat satrli qiymatlarda ishlatishimiz mumkin($regex argument sifatida son qabul qilmaydi)`

## <center> limit

```python
for value in kolleksiya.find().limit(6): #Birinchi 6 ta dokumentni chiqarish
    print(value)
```

```python
query = {"suname":{"$regex":"Y"}}

for value in kolleksiya.find(query).limit(3): # boshidan 3 ta topgan dokumenti chiqaradi
    print(value)
```

## <center> sort

```python
for value in kolleksiya.find().sort("name",1): #sort("name", 1) orqali alfavit bo'yicha sort qilamiz
    print(value)
# Lekin e'tibor bering
# sortlashda agar bizda qaysidur dokumentda kerakli kalit bo'lmasa(masalan name) shu dokumentlar birinchi chiqariladi
# undan keyin esa kalit bo'yicha sort qilib chiqariladi
```

```python
for value in kolleksiya.find().sort("name", -1): #sort("name", -1) orqali alfavit bo'yicha teskari sort qilamiz
    print(value)
```
